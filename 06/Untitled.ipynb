```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.cvtColor(cv2.imread("pies.jpg"), cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
R = img[:,:,0]
R.shape
```

```python
U, D, V = np.linalg.svd(R, full_matrices=False)
```

```python
U.shape
```

```python
D.shape
```

```python
D = np.diag(D)
D.shape
```

```python
V.shape
```

```python
plt.imshow(D)
```

```python
plt.imshow(U@D@V, cmap="gray")
```

```python
k_values = [1, 2, 4, 8, 10, 15, 30, 50, 100]
for k in k_values:
    plt.title(f'k={k}')
    plt.imshow(U[:, :k]@D[:k, :k]@V[:k, :], cmap="gray")
    plt.show()
```

```python
n = 100
k_values = np.arange(1, n+1)
res = np.zeros(n)
for i, k in enumerate(k_values):
    res[i] = sum(sum(pow(U[:, :k]@D[:k, :k]@V[:k, :] - R,2)))
```

```python
plt.plot(k_values, res)
plt.title('Pies')
plt.ylabel('Res')
plt.xlabel('k')
plt.show()
```

## Drugi obraz (stosunkowo prosty)

```python
img2 = cv2.cvtColor(cv2.imread("simp.png"), cv2.COLOR_BGR2RGB)
R = img2[:, :, 0]
U, D, V = np.linalg.svd(R, full_matrices=False)
D = np.diag(D)
plt.imshow(U@D@V, cmap='gray')
```

```python
n = 100
k_values = np.arange(1, n+1)
res = np.zeros(n)
for i, k in enumerate(k_values):
    res[i] = sum(sum(pow(U[:, :k]@D[:k, :k]@V[:k, :] - R,2)))
plt.plot(k_values, res)
plt.title('Symbol')
plt.ylabel('Res')
plt.xlabel('k')
plt.show()
```

```python
k_values = [1, 2, 3, 4, 8, 10, 15, 30]
for k in k_values:
    plt.title(f'k={k}')
    plt.imshow(U[:, :k]@D[:k, :k]@V[:k, :], cmap="gray")
    plt.show()
```
