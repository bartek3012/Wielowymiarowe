```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### Zadanie 1, 2, 3 (var)

```python
D_array = [10, 20, 50, 200, 1000]

for D in D_array:
    s1 = np.random.normal(1, 1, D)
    s2 = np.zeros(D)
    for i in range(D):
        s2[i] = np.random.normal(1, np.log((i+1)), 1)[0]
    s = np.array([s1, s2])
    t = np.arange(1, D+1)

    for i in range(D):
        plt.plot(t, [s[0,i]]*D)
    plt.title(f's1 dla {D} realizacji')
    plt.show()    
    
    for i in range(D):
        plt.plot(t, [s[1,i]]*D)
    plt.title(f's2 dla {D} realizacji')
    plt.show()
    
    plt.plot(t, [s[0].mean()]*D)
    plt.title(f's1 dla redniej z {D} realizacji')
    plt.show()
    
    
    plt.plot(t, [s[1].mean()]*D)
    plt.title(f's2 dla redniej z {D} realizacji')
    plt.show()    
    
    var1 = np.zeros(D)
    for i in range(D):
        var1[i] = s[0, :(i+1)].var()      
    plt.plot(t, var1)
    plt.title(f's1 wariancje')
    plt.show() 
    
    var2 = np.zeros(D)
    for i in range(D):
        var2[i] = s[1, :(i+1)].var()      
    plt.plot(t, var2)
    plt.title(f's2 wariancje')
    plt.show() 
    print('\n')
```

### Zadanie 2

```python
D_array = [20, 50]

for D in D_array:
    len_t = 10000
    t = np.linspace(0, np.pi*4, len_t)
    x1 = np.random.normal(0, 1, D)
    x2 = np.random.normal(1, 0.1, D)
    fi1 = np.random.normal(1, 0.1, D)
    fi2 = np.random.normal(1, 0.1, D)
    
    s1 = np.zeros((D, len_t))
    for i in range(D):   
        s1[i]= x1[i]*np.sin(t +fi1[i])
        
    s2 = np.zeros((D, len_t))
    for i in range(D):   
        s2[i]= x2[i]*np.cos(t +fi2[i])   
    s = np.array([s1, s2])
    

    for i in range(D):
        plt.plot(t, s[0,i])
    plt.title(f's1 dla {D} realizacji')
    plt.show()    
    
    for i in range(D):
        plt.plot(t, s[1,i])
    plt.title(f's2 dla {D} realizacji')
    plt.show()
    
    s1_mean = np.zeros(len_t)
    for t_step in range(len_t):
        s1_mean[t_step] = s[0,:,t_step].mean()
    plt.plot(t, s1_mean)
    plt.title(f's1 dla redniej z {D} realizacji')
    plt.show()
    
    s2_mean = np.zeros(len_t)
    for t_step in range(len_t):
        s2_mean[t_step] = s[1,:,t_step].mean()
    plt.plot(t, s2_mean)
    plt.title(f's2 dla redniej z {D} realizacji')
    plt.show()
   

    plt.plot(t, s[0].var(axis=0))
    plt.title(f's1 dla wariancji z {D} realizacji')
    plt.show()
    
    plt.plot(t, s[1].var(axis=0))
    plt.title(f's2dla wariancji z {D} realizacji')
    plt.show()
    
#     var1 = np.zeros(D)
#     for i in range(D):
#         var1[i] = s[0, :(i+1), 0].var()      
#     plt.plot(t, var1)
#     plt.title(f's1 wariancje')
#     plt.show() 
    
#     var2 = np.zeros(D)
#     for i in range(D):
#         var2[i] = s[1, :(i+1), 0].var()      
#     plt.plot(t, var2)
#     plt.title(f's2 wariancje')
#     plt.show() 
    print('\n')
```
